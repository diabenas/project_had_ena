# Building Model

```python
# Set the number of folds to 5 for validation
k=5
# Create a KFold object with a specified number of divisions
kf = KFold(n_splits=k, shuffle=True, random_state=42)
# Convert feature matrix 'x' and target variable 'y' into DataFrame objects
x_df = pd.DataFrame(x) 
y_df = pd.DataFrame(y)
# Set new column names for the target feature and data frames
new_column_names = ['age','currentSmoker','BPMeds','prevalentHyp','sysBP','diaBP','BMI','heartRate','glucose','TenYearCHD']
x_df.columns = new_column_names
new_column_names =['diabetes']
y_df.columns = new_column_names
# Create empty lists to store evaluation metrics (precision, precision, recall, F1 score)
# and confusion matrices for each cross-validation.
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
confusion_matrices=[]
```

```python
# This code creates multiple machine learning models with different configurations.
# Model 1 is a Support Vector Machine (SVM) classifier with a linear kernel, automatic gamma value,
# regularization parameter C set to 1, and custom class weights specified to handle class imbalance.
model1 = svm.SVC(kernel='linear', gamma='auto', C=1, class_weight={0: 1, 1: 2})
# Model 2 is a Gaussian Naive Bayes classifier, which assumes that features are conditionally independent.
model2 = GaussianNB()
# Model 3 is a Decision Tree classifier with entropy criterion for information gain, a maximum depth of 4,
# and a random state of 42 to ensure reproducibility.
model3 = DecisionTreeClassifier(criterion="entropy",max_depth=4,random_state=42)
# Model 4 is a Logistic Regression classifier, which models the relationship between the features and the target using
# a logistic function to estimate the probabilities of the different classes.
model4=LogisticRegression()
```

```python
# Create empty result lists to store the results of different models
columns = ['SVC','GaussianNB','DecisionTreeClassifier','LogisticRegression', 'backpropagation']
# List to store scale results ( accuracy score)
result1=[]
#List to store scale results ( recall score)
result2=[]
#List to store scale results (f1 score)
result3=[]
# List to store scale results (precision score)
result4=[]
```

```python
# Function to perform model evaluation using cross-validation
def cal(model,x,y):
    k=5
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
   #تقسيم البيانات بإستخدام k-fold
    for train_index, test_index in kf.split(x):
        # the data is split into training and testing sets for each fold of cross-validation.
        x_train, x_test = x.iloc[train_index], x.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        # Feature scaling  performed on the training and testing sets using a 'StandardScaler'.
        scaler = StandardScaler()
        X_train = scaler.fit_transform(x_train)
        X_test = scaler.transform(x_test)
        # Fit the model on the training data and make predictions on the testing data
        model.fit(X_train, y_train)
        pred=model.predict(X_test)
        # Calculate and store evaluation metrics for the fold
        accuracy_scores.append(accuracy_score(pred , y_test)) 
        recall_scores.append(recall_score(pred , y_test)) 
        f1_scores.append(f1_score(pred , y_test))
        precision_scores.append(precision_score(pred , y_test))
        # Calculate and store the confusion matrix for the fold
        cm=confusion_matrix(pred , y_test)
        confusion_matrices.append( cm)
    # Calculate the mean of evaluation metrics across all folds
    mean_accuracy = sum(accuracy_scores) / len(accuracy_scores)
    mean_recall = sum(recall_scores) / len(recall_scores)
    mean_f1 = sum(f1_scores) / len(f1_scores)
    mean_precision = sum(precision_scores) / len(precision_scores)
    # Calculate the mean of confusion matrices across all folds
    con_matrix = sum(confusion_matrices)/5
    # The mean evaluation metrics are stored in the result lists 'result1', 'result2', 'result3', and 'result4'.    
    result1.append(mean_accuracy)
    result2.append(mean_recall)
    result3.append(mean_f1)
    result4.append(mean_precision)
    # Visualize the mean confusion matrix using a heatmap
    sns.heatmap(   con_matrix , annot=True)
    # The model, along with the mean evaluation metrics and the mean confusion matrix, printed.
    print(model)
    print('accuracy is : ', mean_accuracy*100 ,'recall is :', mean_recall*100,'f1 is :', mean_f1*100,'precision is :', mean_precision*100)
    # DataFrame is created to display the mean confusion matrix.
    cm_df = pd.DataFrame(con_matrix, index=['Actual N', 'Actual P'], columns=['Predicted N', 'Predicted P'])
    print(cm_df) 
# The function is called using the form SVM
```

```python
# The model structure consists of an 8-unit dense layer and the sigmoid activation function as the input layer,
# Followed by a single unit dense layer and the sigmoid activation function as the output layer
# The model is assembled using the “Adam” optimizer and the “binary_crossentropy” loss function for binary classification,
# And “accuracy” as a measure of monitoring during training.
# Define the architecture of the model
model5 = keras.Sequential([
    keras.layers.Dense(8, activation='sigmoid', input_shape=(10,)),
    keras.layers.Dense(1, activation='sigmoid')
])
#Define the objective function
model5.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
for train_index, test_index in kf.split(x):
    x_train, x_test = x_df.iloc[train_index], x_df.iloc[test_index]
    y_train, y_test = y_df.iloc[train_index], y_df.iloc[test_index]

        # Implement feature scaling on training and test sets
    scaler = StandardScaler()
    X_train = scaler.fit_transform(x_train)
    X_test = scaler.transform(x_test)

    # Train the model on training data for a specified number of epochs
    model5.fit(X_train, y_train, epochs=100,batch_size=128, verbose=0)

    # Make predictions on test data
    y_pred = (model5.predict(X_test) > 0.5).astype(int)
    
    accuracy_scores.append(accuracy_score(y_pred , y_test)) 
    recall_scores.append(recall_score(y_pred , y_test)) 
    f1_scores.append(f1_score(y_pred , y_test))
    precision_scores.append(precision_score(y_pred , y_test))
    confusion_matrices.append( confusion_matrix(y_pred , y_test))

mean_accuracy = sum(accuracy_scores) / len(accuracy_scores)
mean_recall = sum(recall_scores) / len(recall_scores)
mean_f1 = sum(f1_scores) / len(f1_scores)
mean_precision = sum(precision_scores) / len(precision_scores)
con_matrix = sum(confusion_matrices)/5
         
result1.append(mean_accuracy)
result2.append(mean_recall)
result3.append(mean_f1)
result4.append(mean_precision)

sns.heatmap(   con_matrix , annot=True)
#print(model5)
print(model5)
print('accuracy is : ', mean_accuracy*100 ,'recall is :', mean_recall*100,'f1 is :', mean_f1*100,'precision is :', mean_precision*100)

cm_df = pd.DataFrame(con_matrix, index=['Actual N', 'Actual P'], columns=['Predicted N', 'Predicted P'])
print(cm_df)
        
```

```python
# Plot the model architecture
plot_model(model5, to_file='model_architecture.png', show_shapes=True)# Create a directed graph     age,currentSmoker,BPMeds,prevalentHyp,sysBP,diaBP,BMI,heartRate,glucose,TenYearCHD
G = nx.DiGraph()

# Add input layer nodes
input_layer = ['age', 'currentSmoker', 'BPMeds', 'prevalentHyp', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose', 'TenYearCHD']
G.add_nodes_from(input_layer, layer=0)

# Add hidden layer nodes
hidden_layer = ['Hidden 1', 'Hidden 2', 'Hidden 3', 'Hidden 4', 'Hidden 5', 'Hidden 6', 'Hidden 7', 'Hidden 8']
G.add_nodes_from(hidden_layer, layer=1)

# Add output layer node
output_layer = ['diabetes']
G.add_nodes_from(output_layer, layer=2)

# Connect input layer to hidden layer with weights
input_hidden_weights = model5.layers[0].get_weights()[0]
for i, input_node in enumerate(input_layer):
    for j, hidden_node in enumerate(hidden_layer):
        weight = input_hidden_weights[i][j]
        G.add_edge(input_node, hidden_node, weight=weight)

# Connect hidden layer to output layer with weights
hidden_output_weights = model5.layers[1].get_weights()[0]
for i, hidden_node in enumerate(hidden_layer):
    weight = hidden_output_weights[i][0]
    G.add_edge(hidden_node, 'diabetes', weight=weight)

# Set positions for node labels
pos = nx.multipartite_layout(G, subset_key='layer', align='vertical')

# Draw the neural network
plt.figure(figsize=(10, 6))
nx.draw_networkx(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=10, font_weight='bold')

# Draw edge labels with weights
edge_labels = nx.get_edge_attributes(G, 'weight')
edge_labels = {k: f"{v:.3f}" for k, v in edge_labels.items()}  # Format weights to three decimal places
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, label_pos=0.5, rotate=False)

plt.title('Neural Network Architecture')
plt.axis('off')
plt.show()
```
